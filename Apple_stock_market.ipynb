```python
# prediction of closing prices of Apple stock market
# Using Decision tree algorithm

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('AAPL.csv')
df.head()
```

```python
newdf =df.drop(0)
data= newdf.reset_index(drop=True)
data.head()
```

```python
X = data[['Open', 'High', 'Low', 'Volume']]
Y = data[['Close','Adj Close']]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
X_train
X_test
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion='squared_error')
regressor.fit(X_train, Y_train)
```

```python
regressor.predict(X_test)
```

```python
regressor.score(X_test, Y_test)
```

```python
X = data[['Open', 'High', 'Low', 'Volume']]
X.head()
```

```python
regressor.predict([[176.98, 179.10, 176.55, 45212000]])
```

```python
regressor.score(X, Y)
```

```python
predicted_prices = regressor.predict([[176.98, 179.10, 176.55, 45212000]])
predicted_close = predicted_prices[0][0]
predicted_adj_close = predicted_prices[0][1]

print(f"Predicted Close: {predicted_close}")
print(f"Predicted Adj Close: {predicted_adj_close}")
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))
plot_tree(regressor, filled=True, feature_names=X.columns, rounded=True)
plt.show()
```
